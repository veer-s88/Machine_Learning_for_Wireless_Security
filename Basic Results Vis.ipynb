```python
import numpy as np
import pandas as pd
```

```python
dcf_sim = pd.read_csv('simulation_results.csv')
dcf_sim.head()
```

```python
dcf_sim.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
dcf_sim.set_index('number of nodes', inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
plt.plot(dcf_sim['collision probability'], 'o-')
plt.xlabel('number of nodes')
plt.ylabel('collision probability')
plt.title('Collision Probability per Packet vs No. of Nodes in Network')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(dcf_sim['total throughput'], 'o-')
plt.xlabel('number of nodes')
plt.ylabel('total throughput (b/s)')
plt.title('Total Throughput across Network vs No. of Nodes in Network')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(dcf_sim['average delay'], 'o-')
plt.xlabel('number of nodes')
plt.ylabel('average delay (s)')
plt.title('Average delay per Packet across Network vs No. of Nodes in Network')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(dcf_sim['simulation time'], 'o-')
plt.xlabel('number of nodes')
plt.ylabel('simulation time (s)')
plt.title('Total Simulation Time for 100,000 Contention Rounds vs No. of Nodes in Network')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(dcf_sim['dropped packet'], 'o-')
plt.xlabel('number of nodes')
plt.ylabel('dropped packets')
plt.title('Total Number of Dropped Packets vs No. of Nodes in Network')
plt.show()
```
