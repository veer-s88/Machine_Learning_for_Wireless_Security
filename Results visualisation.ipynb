```python
import numpy as np
import pandas as pd
```

```python
dcf_sim = pd.read_csv('simulation_results.csv')
dcf_sim_rogue_1 = pd.read_csv('simulation_results_rogue_node1.csv')
dcf_sim_rogue_2 = pd.read_csv('simulation_results_rogue_node2.csv')
dcf_sim_rogue_3 = pd.read_csv('simulation_results_rogue_node3.csv')
dcf_sim_rogue_4 = pd.read_csv('simulation_results_rogue_node4.csv')
dcf_sim_rogue_5 = pd.read_csv('simulation_results_rogue_node5.csv')
```

```python
dcf_sim.tail()
```

```python
dcf_sim_rogue_3.head()
```

```python
dcf_sim_rogue_2.head()
```

```python
dcf_sim.set_index('number of nodes', inplace=True)
dcf_sim_rogue_1.set_index('number of nodes', inplace=True)
dcf_sim_rogue_2.set_index('number of nodes', inplace=True)
dcf_sim_rogue_3.set_index('number of nodes', inplace=True)
dcf_sim_rogue_4.set_index('number of nodes', inplace=True)
dcf_sim_rogue_5.set_index('number of nodes', inplace=True)
```

```python
dcf_sim.drop(['Unnamed: 0'], axis=1, inplace=True)
dcf_sim_rogue_1.drop(['Unnamed: 0'], axis=1, inplace=True)
dcf_sim_rogue_2.drop(['Unnamed: 0'], axis=1, inplace=True)
dcf_sim_rogue_3.drop(['Unnamed: 0'], axis=1, inplace=True)
dcf_sim_rogue_4.drop(['Unnamed: 0'], axis=1, inplace=True)
dcf_sim_rogue_5.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
dcf_sim_rogue_5.head()
```

```python
dcf_sim_rogue_1.columns = ['collision_probability_rogue_1', 'total_throughput_rogue_1', 'average_delay_rogue_1', 'simulation_time_rogue_1','dropped_packet_rogue_1']
dcf_sim_rogue_2.columns = ['collision_probability_rogue_2', 'total_throughput_rogue_2', 'average_delay_rogue_2', 'simulation_time_rogue_2','dropped_packet_rogue_2']
dcf_sim_rogue_3.columns = ['collision_probability_rogue_3', 'total_throughput_rogue_3', 'average_delay_rogue_3', 'simulation_time_rogue_3','dropped_packet_rogue_3']
dcf_sim_rogue_4.columns = ['collision_probability_rogue_4', 'total_throughput_rogue_4', 'average_delay_rogue_4', 'simulation_time_rogue_4','dropped_packet_rogue_4']
dcf_sim_rogue_5.columns = ['collision_probability_rogue_5', 'total_throughput_rogue_5', 'average_delay_rogue_5', 'simulation_time_rogue_5','dropped_packet_rogue_5']
```

```python
join_table = pd.concat([dcf_sim, dcf_sim_rogue_1, dcf_sim_rogue_2, dcf_sim_rogue_3, dcf_sim_rogue_4, dcf_sim_rogue_5], axis=1)
join_table.head(6)
```

```python
col_prob = join_table[['collision probability', 'collision_probability_rogue_1', 'collision_probability_rogue_2', 'collision_probability_rogue_3', 'collision_probability_rogue_4', 'collision_probability_rogue_5']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(col_prob, marker='o')
plt.title('Comparison of fair wireless network and network with cheating nodes: collision probability')
plt.xlabel('number of nodes')
plt.ylabel('collision probability')
plt.legend(['Fair wireless network', '1 cheating node', '2 cheating nodes', '3 cheating nodes', '4 cheating nodes', '5 cheating nodes'])
plt.show()
```

```python
tot_thruput = join_table[['total throughput', 'total_throughput_rogue_1', 'total_throughput_rogue_2', 'total_throughput_rogue_3', 'total_throughput_rogue_4', 'total_throughput_rogue_5']]
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(tot_thruput, marker='o')
plt.title('Comparison of fair wireless network and network with cheating nodes: total network throughput')
plt.xlabel('number of nodes')
plt.ylabel('total network throughput (b/s)')
plt.legend(['Fair wireless network', '1 cheating node', '2 cheating nodes', '3 cheating nodes', '4 cheating nodes', '5 cheating nodes'])
plt.show()
```

```python
avg_delay = join_table[['average delay', 'average_delay_rogue_1', 'average_delay_rogue_2', 'average_delay_rogue_3', 'average_delay_rogue_4', 'average_delay_rogue_5']]
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(avg_delay, marker='o')
plt.title('Comparison of fair wireless network and network with cheating nodes: average delay')
plt.xlabel('number of nodes')
plt.ylabel('average delay (s)')
plt.legend(['Fair wireless network', '1 cheating node', '2 cheating nodes', '3 cheating nodes', '4 cheating nodes','5 cheating nodes'])
plt.show()
```

```python
sim_time = join_table[['simulation time','simulation_time_rogue_1', 'simulation_time_rogue_2', 'simulation_time_rogue_3', 'simulation_time_rogue_4', 'simulation_time_rogue_5']]
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(sim_time.loc[5:], marker='o')
plt.title('Comparison of working network and network with cheating nodes: simulation time')
plt.xlabel('number of nodes')
plt.ylabel('simulation time (s)')
plt.legend(['Working wireless network', '1 rogue node', '2 rogue nodes', '3 rogue nodes', '4 rogue nodes','5 rogue nodes'])
plt.show()
```

```python
pckt_drop = join_table[['dropped packet', 'dropped_packet_rogue_1', 'dropped_packet_rogue_2', 'dropped_packet_rogue_3','dropped_packet_rogue_4', 'dropped_packet_rogue_5']]
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(pckt_drop, marker='o')
plt.title('Comparison of working network and network with cheating nodes: dropped packets')
plt.xlabel('number of nodes')
plt.ylabel('Dropped Packets')
plt.legend(['Working wireless network', '1 rogue node', '2 rogue nodes', '3 rogue nodes', '4 rogue nodes','5 rogue nodes'])
plt.show()
```
